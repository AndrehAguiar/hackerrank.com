### Coumpound Event Probabity

```python
from fractions import Fraction

x = [4, 3]
y = [5, 4]
z = [4, 4]

# The probabilities of pulling red form each urn
xr = x[0] / sum(x)
yr = y[0] / sum(y)
zr = z[0] / sum(z)

p = (xr * yr * (1 - zr)) + (xr * (1 - yr) * zr) + ((1 - xr) * yr * zr)
p_frac = Fraction(p).limit_denominator() # Fraction form

print(f"The probability of drawing 2 reds and one black is {p}, or {p_frac}.")
```

### Binomial Distribution
The ratio of boys to girls for babies born in Russia is . If there is  child born per birth, what proportion of Russian families with exactly  children will have at least  boys?

P(x = 3) = P(x = 3) + P(x = 4) + P(x = 5) + P(x = 6)

```python
p = 1.09/(1+1.09)
k = 3
n = 6
result = 0
def getFactorial(int):
    b = int
    i = b-1
    while i > 0:
        b *= i
        i-=1
    return(b)

for i in range(k,n+1):
    result += getFactorial(n) / (getFactorial(i)*getFactorial(n-i if i < n else 1)) * (p**i) * ((1-p)**(n-i))
    
print("%.3f" % result)
```

### Pearson Correlation Coefficient

<hr />

Given two <i><b>n</b></i>-element data sets, <i><b>X</b></i> and <i><b>Y</b></i>, calculate the value of the Pearson correlation coefficient.



```python
n = 10

X = list(map(float, '10 9.8 8 7.8 7.7 7 6 5 4 2'.split()))
Y = list(map(float, '200 44 32 24 22 17 15 12 8 4'.split()))

x_mean = sum(X)/n
y_mean = sum(Y)/n

x_std = (sum([(i-(sum(X)/n))**2 for i in X])/n)**0.5
y_std = (sum([(i-(sum(Y)/n))**2 for i in Y])/n)**0.5

corr = 0
for i in range(n):
    corr += (float(X[i]) - x_mean) * (float(Y[i]) - y_mean)

corr = corr / (n * (x_std * y_std))
print('%.3f'%corr)
```


Given two <i><b>n</b></i>-element data sets, <i><b>X</b></i> and <i><b>Y</b></i>, calculate the value of Spearman's rank correlation coefficient.

```python
n = 10

X = list(map(float, '10 9.8 8 7.8 7.7 1.7 6 5 1.4 2'.split()))
Y = list(map(float, '200 44 32 24 22 17 15 12 8 4'.split()))

rx = [n - sorted(X, reverse = True).index(x) for i, x in enumerate(X)]
ry = [n - sorted(Y, reverse = True).index(y) for i, y in enumerate(Y)]
di = [(rx[i] - ry[i]) ** 2 for i in range(n)]

r = 1 - (6 * sum(di)) / (n*((n**2) - 1))

print('%.3f'%r)
```

A group of five students enrolls in Statistics immediately after taking a Math aptitude test. Each student's Math aptitude test score, , and Statistics course grade, <i><b>x</b></i>, can be expressed as the following list of <i><b>(x, y)</b></i> points:

If a student scored an <i><b>80</b></i> on the Math aptitude test, what grade would we expect them to achieve in Statistics? Determine the equation of the best-fit line using the least squares method, then compute and print the value of <i><b>y</b></i> when <i><b>x = 80</b></i>.

#### Using the Pearson Correlation Coefficient

```python
xy = [(95, 85),(85, 95),(80, 70),(70, 65),(60, 70)]
x = [x[0] for i, x in enumerate(xy)]
y = [y[1] for i, y in enumerate(xy)]

x_mean = sum(x) / len(x)
y_mean = sum(y) / len(y)

x_std = (sum([(i-(x_mean))**2 for i in x])/len(x))**0.5
y_std = (sum([(i-(y_mean))**2 for i in y])/len(y))**0.5
    
corr = 0
for i in range(len(x)):
    corr += (float(x[i]) - x_mean) * (float(y[i]) - y_mean)

corr = corr / (len(x) * (x_std * y_std))

b = corr * (y_std / x_std)
a = y_mean - (b * x_mean)

Y = a + b * x[2]

print('%.3f'%Y)
```

#### Simple Regression Line

```python
#xy = [(95, 85),(85, 95),(80, 70),(70, 65),(60, 70)]
#X = [x[0] for x, x in enumerate(xy)]
#Y = [y[1] for i, y in enumerate(xy)]
X, Y = zip(*[(x, y) for x, y in xy])
n = len(X)

sum_X = sum(X)
mean_X = sum_X / n

sum_Y = sum(Y)
mean_Y = sum_Y / n

X2 = sum([x**2 for x in X])
XY = sum([X[i] * Y[i] for i in range(n)])

b = ((n * XY) - (sum_X * sum_Y)) / ((n * X2) - (sum_X ** 2))

a = mean_Y - b * mean_X
y = a + b * 80
print('%.3f'%y)
```
